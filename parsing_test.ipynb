### pdfplumber

```python
import io
import requests

url = 'https://www.lachambre.be/FLWB/PDF/55/3496/55K3496001.pdf'
# def extract_pdf_by_url(url):
rq = requests.get(url)

left_column_text_all_pages = ""
right_column_text_all_pages = ""

with pdfplumber.open(io.BytesIO(rq.content)) as pdf:
    for page in pdf.pages:
        words = page.extract_words()

        threshold_x = 300

        left_column_words = [word for word in words if word['x0'] < threshold_x]
        right_column_words = [word for word in words if word['x0'] >= threshold_x]

        left_column_text = ' '.join([word['text'] for word in left_column_words])
        right_column_text = ' '.join([word['text'] for word in right_column_words])

        left_column_text_all_pages += left_column_text + '\n'
        right_column_text_all_pages += right_column_text + '\n'
   

print("Left Column Text from All Pages:")
print(left_column_text_all_pages)

print("Right Column Text from All Pages:")
print(right_column_text_all_pages)
```

```python
import re

pattern = r'\w+-\s+\w+'

matches = re.findall(pattern, left_column_text_all_pages)
print(matches)

left_column_text_all_pages = re.sub(r'(?<=\w)-(\s)(?=\w)', r'', left_column_text_all_pages)

```

```python
print(left_column_text_all_pages)
```

```python
pattern = r'\w+-\s+\w+'

matches = re.findall(pattern, right_column_text_all_pages)
print(matches)

right_column_text_all_pages = re.sub(r'(?<=\w)-(\s)(?=\w)', r'', right_column_text_all_pages)
print(right_column_text_all_pages)
```
